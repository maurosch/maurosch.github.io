```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("casos_covid19.csv")
display(df)
print(df.dtypes)
```

```python
df[(df['fecha_fallecimiento'].notnull())]
```

```python
data = df[(df['fecha_fallecimiento'].notnull()) & (df['edad'] >= 20) & (df['edad'] <= 100)].groupby(['edad']).size()
```

```python
plt.bar(data.index,data.values)
plt.grid
plt.xlim([20, 100])
plt.xlabel('Edad')
plt.ylabel('Cantidad fallecidos')
plt.savefig("assets/fallecimiento.svg")
plt.show()
```

```python
np.sum(data.values)
```

```python
data2 = df[(df['edad'] >= 20) & (df['edad'] <= 100) & (df['clasificacion'] == 'confirmado')].groupby(['edad']).size()
plt.bar(data.index,data.values/data2.values)
plt.grid()
plt.xlim([20, 100])
plt.xlabel('Edad')
plt.ylabel('Porcentaje fallecidos de casos confirmados')
plt.savefig("assets/prob-fallecimiento.svg")
plt.show()
```

```python
df['year'] = df['fecha_clasificacion'].apply(lambda v : int(v[5:9]))
```

```python
data = df[(df['year'] == 2020) & (df['fecha_fallecimiento'].notnull()) & (df['edad'] >= 20) & (df['edad'] <= 100)].groupby(['edad']).size()
data2 = df[(df['year'] == 2020) & (df['edad'] >= 20) & (df['edad'] <= 100) & (df['clasificacion'] == 'confirmado')].groupby(['edad']).size()
plt.bar(data.index,data.values/data2.values)

data = df[(df['year'] == 2021) & (df['fecha_fallecimiento'].notnull()) & (df['edad'] >= 20) & (df['edad'] <= 100)].groupby(['edad']).size()
data2 = df[(df['year'] == 2021) & (df['edad'] >= 20) & (df['edad'] <= 100) & (df['clasificacion'] == 'confirmado')].groupby(['edad']).size()
plt.bar(data.index,data.values/data2.values)
plt.grid()
plt.xlim([20, 100])
plt.xlabel('Edad')
plt.ylabel('Porcentaje fallecidos de casos confirmados')
plt.legend(['2020','2021'])
plt.savefig("assets/prob-fallecimiento-2020-2021.svg")
plt.show()
```

```python
data = df[(df['clasificacion'] == 'confirmado')].groupby(['edad']).size()
plt.bar(data.index,data.values)
plt.grid()
plt.xlim([20, 100])
plt.xlabel('Edad')
plt.ylabel('Cantidad contagiados')
plt.savefig("assets/contagiados.svg")
plt.show()
```

```python
df['fecha_clasificacion_formatted'] = df['fecha_clasificacion'].apply(lambda v : datetime.strptime(v,'%d%b%Y:00:00:00.000000'))
```

```python
data = df.groupby(df['fecha_clasificacion_formatted']).size()
plt.bar(data.index,data.values)
plt.grid()
plt.xlabel('Fecha')
plt.xlim([datetime(2019,12,1), datetime(2021,7,1)])
plt.ylabel('Cantidad contagiados')
plt.savefig("assets/casos-positivos-en-el-tiempo.svg")
plt.show()
```
